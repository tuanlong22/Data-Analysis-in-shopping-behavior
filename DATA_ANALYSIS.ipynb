###DATA VISUALIZATION###

```python
# Exploring gender in this data
gender = df['Gender'].value_counts()
colors = ['blue','yellow']
plt.bar(['Female','Male'],(gender[0],gender[1]), color = colors)
plt.title('Gender in data')
plt.ylabel('Numbers')
plt.xlabel('Genders')
```

From the bar chart above, we conclude that the number of women doing shopping is more than men





```python
# Explain people doing shopping by age and gender
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
# Using a for loop to give the bars exact values
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total person doing shopping by age group
ax = sns.countplot(data = df, x='Age Group')
for bars in ax.containers:
  ax.bar_label(bars)
```

From the chart above, we easily see that the most buyers is around 26 to 35

```python
# Total numbers of orders of top 10 states
order_state = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False)
print(order_state)
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.barplot(data = order_state, x = 'State', y = 'Orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
# Occupation
# How occupation affect buying behaviours
occupation = df.groupby(['Occupation'], as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False)
print(occupation)
```

```python
ax = sns.barplot(data = occupation, x = 'Occupation', y ='Orders')
for bars in ax.containers:
  ax.bar_label(bars)
```

From the chart above, we conclude that IT is the king of jobs

```python
# Product Category
sns.set(rc={'figure.figsize':(20,5)})
category = df.groupby(['Product_Category'], as_index = False).sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = category, x = 'Product_Category',y= 'Amount')
```

So, food is type of product with most bought
